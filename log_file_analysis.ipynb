```python
# parse the log file and save as .parquet format
adv.logs_to_df(log_file='access_log',
               output_file='access_log.parquet',
               errors_file='access.txt',
               log_format='([^-]*) ([^-]*) ([^ ]*) ([^ ]*) \[([^]]*)\] "([^"]*)" ([^ ]*) ([^ ]*) "([^"]*)" "([^"]*)"',
               fields=['host','client','userid','','datetime', 'request', 'size','status','referer','user_agent']
               )
```

```python
logs_df = pd.read_parquet('access_log.parquet')

logs_df['datetime'] = pd.to_datetime(logs_df['datetime'], format='%d/%b/%Y:%H:%M:%S %z')

logs_df
```

```python
# prompt: create a new dataframe from logs_df and remove GET for an image where the URL is pass through, all bots and anything on localhost as the client, POST /wp-cron.php or admin-ajax.php

import pandas as pd

# Assuming logs_df is already defined from the previous code

def process_logs(logs_df):
    # Create a copy to avoid modifying the original DataFrame
    processed_logs = logs_df.copy()

    # Conditions for removal
    conditions = [
        processed_logs['request'].str.contains(r'^GET .* \.(?:jpg|jpeg|png|gif|bmp|ico)$', regex=True), #images
        processed_logs['client'].str.contains(r'bot|crawler|spider', case=False, regex=True), # bots
        processed_logs['client'].str.startswith('127.0.0.1'), # localhost
        processed_logs['request'].str.contains(r'^POST /wp-cron\.php', regex=True),
        processed_logs['request'].str.contains(r'^POST /admin-ajax\.php', regex=True),
        processed_logs['user_agent'].str.contains(r'^WordPress/', regex=True),
        processed_logs['user_agent'].str.contains(r'^Python/', regex=True)
      ]

    # Combine all conditions
    processed_logs = processed_logs[~pd.concat(conditions, axis=1).any(axis=1)]

    return processed_logs

# Apply the function to the logs_df DataFrame
new_logs_df = process_logs(logs_df)
new_logs_df
```

```python
# prompt: export new_logs_df to csv

new_logs_df.to_csv('new_logs.csv', index=False)
```

```python
request_url_df = adv.url_to_df(new_logs_df['request'])
request_url_df = request_url_df.add_prefix('request_')
request_url_df.head(10)
```

```python
# prompt: expport to csv

request_url_df.to_csv('request_url.csv', index=False)
```

```python
referer_url_df = adv.url_to_df(new_logs_df['referer'])
referer_url_df = referer_url_df.add_prefix('referer_')
referer_url_df.head(10)
```

```python
ua_df = pd.json_normalize([user_agent_parser.Parse(ua) for ua in new_logs_df['user_agent']])
ua_df.columns = 'ua_' + ua_df.columns.str.replace('user_agent\.', '', regex=True)
ua_df.head(10)
```

```python
(pd.concat([new_logs_df, request_url_df, referer_url_df, ua_df], axis=1)
.to_parquet('data/adv_logs_final.parquet', index=False, version='2.4'))
```
